```python
import matplotlib.pyplot as plt 
%matplotlib inline

plt.rcParams['figure.dpi'] = 250
plt.rcParams['savefig.dpi'] = 250
```

```python
# open cylinder

P = [1.5 , 2 , 2.5]

Eps_H_theo_open = [172.2, 258.2,	344.2,	430.3 , 516.2]
Eps_H_exp_open = [316,	404,	492]



plt.plot([1 , 1.5 , 2 , 2.5 , 3] , Eps_H_theo_open , color = "b" , label = "Theoretical")
plt.scatter(P , Eps_H_exp_open , color = "r", label = "Experimental")

plt.xticks(P)
plt.xlim(0.5,3.5)
plt.xlabel("P(MPa)")
plt.ylabel("Epsilon_H(x 10^-6 m)")
plt.ylim(100 , 550)
plt.legend(loc = "upper left")
plt.title("Figure 1: Epsiolon_H VS. Pressure(P) Curve in the Open-End Cylinder" , fontsize = 10)


plt.tight_layout()
plt.show()



```
