```python
F = [0 , 5,10,15,20 , 25]
Delta_theo = [0 , 0.1311,	0.2622,	0.3933,	0.5244 , 0.6555]
Delta_exp = [0.13,	0.26,	0.4,	0.57]

plt.plot(F , Delta_theo , color = "b" , label = "Theoretical")
plt.scatter([5,10,15,20] , Delta_exp , color = "r", label = "Experimental")

plt.xticks(F)
plt.xlim(0,27)
plt.xlabel("F(N)")
plt.ylabel("Delta(mm)")
plt.ylim(0 , 0.7)
plt.legend(loc = "upper left")
plt.title("Figure 1: Deflection VS. Force Curve")



plt.show()
```

```python
import numpy as np
import numpy as np
from scipy.interpolate import make_interp_spline


L = np.array([300,	400,	500,	600])
Delta_theo = np.array([0.0566,	0.1342,	0.2622,	0.4531])


X_Y_Spline = make_interp_spline(L, Delta_theo)
 
# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(L.min()-50, L.max()+30, 500)
Y_ = X_Y_Spline(X_)
 
# Plotting the Graph
Delta_exp = [0.06,	0.16,	0.27,	0.47]

plt.plot(X_ , Y_ , color = "b" , label = "Theoretical")
plt.scatter(L , Delta_exp , color = "r", label = "Experimental")

plt.xticks(L)
plt.xlim(200,700)
plt.xlabel("L(mm)")
plt.ylabel("Delta(mm)")
plt.ylim(0 , 0.6)
plt.legend(loc = "upper left")
plt.title("Figure 2: Deflection VS. Beam Length Curve")
```

```python
b = np.array([15,	20,	25,	30])
Delta_theo = np.array([0.7374,	0.553,	0.4425,	0.3686])


X_Y_Spline = make_interp_spline(b, Delta_theo)
 
# Returns evenly spaced numbers
# over a specified interval.
X_ = np.linspace(b.min()-2, b.max()+2, 500)
Y_ = X_Y_Spline(X_)
 
# Plotting the Graph
Delta_exp = [0.85,	0.64,	0.51,	0.42]

plt.plot(X_ , Y_ , color = "b" , label = "Theoretical")
plt.scatter(b , Delta_exp , color = "r", label = "Experimental")

plt.xticks(b)
plt.xlim(10,35)
plt.xlabel("b(mm)")
plt.ylabel("Delta(mm)")
plt.ylim(0 , 1)
plt.legend(loc = "upper right")
plt.title("Figure 3: Deflection VS. Beam Width(b) Curve")
plt.show()
```

```python
h = np.array([3,	4,	6, 7,	8])
Delta_theo = np.array([1.0487,	0.4424,	0.115, 0.078,	0.06])

from scipy.interpolate import make_interp_spline, BSpline

# 300 represents number of points to make between T.min and T.max
xnew = np.linspace(h.min()-0.15, h.max()+0.3, 300) 

spl = make_interp_spline(h, Delta_theo, k=3)  # type: BSpline
power_smooth = spl(xnew)
 
plt.plot(xnew, power_smooth , color = "b")




# Plotting the Graph
Delta_exp = [1.16,	0.5,	0.116,	0.06]

plt.scatter([3,4,6,8] , Delta_exp , color = "r", label = "Experimental")

plt.xticks(h)
plt.xlim(1,10)
plt.xlabel("h(mm)")
plt.ylabel("Delta(mm)")
plt.ylim(0 , 1.3)
plt.legend(loc = "upper right")
plt.title("Figure 4: Deflection VS. Beam Thickness(h) Curve")
plt.show()
```
